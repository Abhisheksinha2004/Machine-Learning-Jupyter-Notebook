```python
# Function Transformer->to convert non normal distribution data to normal distribution data 
```

```python
# we can use this functiontransformer when there is important data present in outlier,the 
# functiontransformer will not remove outlier and transform the data to normal distribution  
```

```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset=pd.read_csv(r"C:\Users\HP\Downloads\loan (1).csv")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
sns.distplot(dataset["CoapplicantIncome"])
plt.show()
```

```python
# using outlier detection method
```

```python
q1=dataset["CoapplicantIncome"].quantile(0.25)
q3=dataset["CoapplicantIncome"].quantile(0.75)
IQR=q3-q1
```

```python
min_r=q1-(1.5*IQR)
max_r=q3+(1.5*IQR)
min_r,max_r
```

```python
dataset=dataset[dataset["CoapplicantIncome"]<=max_r]
```

```python
sns.distplot(dataset["CoapplicantIncome"])
plt.show()
```

```python
from sklearn.preprocessing import FunctionTransformer
```

```python
ft=FunctionTransformer(func=np.log1p)
```

```python
#we can not use log base 10 or e as our data may be zero or negative which will result in abnormal behavoiur
```

```python
ft.fit(dataset[["CoapplicantIncome"]])
```

```python
dataset["CoapplicantIncome_tf"]=ft.transform(dataset[["CoapplicantIncome"]])
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
sns.distplot(dataset["CoapplicantIncome"])
plt.title("Before")

plt.subplot(1,2,2)
sns.distplot(dataset["CoapplicantIncome_tf"])
plt.title("After")
plt.show()
```

```python
dataset=pd.read_csv(r"C:\Users\HP\Downloads\loan (1).csv")
dataset.head(3)
```

```python
sns.distplot(dataset["CoapplicantIncome"])
plt.show()
```

```python
from sklearn.preprocessing import FunctionTransformer
```

```python
ft=FunctionTransformer(func=np.log1p)
```

```python
ft.fit(dataset[["CoapplicantIncome"]])
```

```python
dataset["CoapplicantIncome_tf"]=ft.transform(dataset[["CoapplicantIncome"]])
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
sns.distplot(dataset["CoapplicantIncome"])
plt.title("Before")

plt.subplot(1,2,2)
sns.distplot(dataset["CoapplicantIncome_tf"])
plt.title("After")
plt.show()
```

```python
ft1=FunctionTransformer(func=lambda x : x**2)
ft1.fit(dataset[["CoapplicantIncome"]])
dataset["CoapplicantIncome_tf1"]=ft1.transform(dataset[["CoapplicantIncome"]])
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
sns.distplot(dataset["CoapplicantIncome"])
plt.title("Before")

plt.subplot(1,2,2)
sns.distplot(dataset["CoapplicantIncome_tf1"])
plt.title("After")
plt.show()
```
