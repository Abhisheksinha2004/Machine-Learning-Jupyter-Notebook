```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Feature Scaling (Standardization)
```

```python
dataset=pd.read_csv(r"C:\Users\HP\Downloads\loan (1).csv")
dataset.head(3)
```

```python
dataset.isnull().sum() 
```

```python
sns.distplot(dataset["ApplicantIncome"])
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
ss.fit(dataset[["ApplicantIncome"]])
```

```python
ss.transform(dataset[["ApplicantIncome"]])
```

```python
dataset["ApplicantIncome_ss"]=pd.DataFrame(ss.transform(dataset[["ApplicantIncome"]]),columns=["x"])
```

```python
dataset.head(3)
```

```python
dataset.describe()
```

```python
sns.distplot(dataset["ApplicantIncome_ss"])
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(dataset["ApplicantIncome"])

plt.subplot(1,2,2)
plt.title("After")
sns.distplot(dataset["ApplicantIncome_ss"])
plt.show()
```

```python
#Feature Scaling (Normalization)
dataset=pd.read_csv(r"C:\Users\HP\Downloads\loan (1).csv")
dataset.head(3)
```

```python
dataset.describe()
```

```python
sns.distplot(dataset["CoapplicantIncome"])
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
ms=MinMaxScaler()
ms.fit(dataset[["CoapplicantIncome"]])
```

```python
ms.transform(dataset[["CoapplicantIncome"]])
```

```python
dataset["CoapplicantIncome_min"]= ms.transform(dataset[["CoapplicantIncome"]])
```

```python
dataset.head(3)
```

```python
#transform returns a 2D numpy array , to get it convert into pandas dataframe we will use 
dataset["CoapplicantIncome_min"]= ms.transform(dataset[["CoapplicantIncome"]]).ravel()
```

```python
dataset.head(3)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(dataset["CoapplicantIncome"])

plt.subplot(1,2,2)
plt.title("After")
sns.distplot(dataset["CoapplicantIncome_min"])

plt.show()
```
