```python
import numpy
print(numpy.__version__)
```

```python
import pandas as pd
from mlxtend.feature_selection import SequentialFeatureSelector
```

```python
dataset=pd.read_csv(r"C:\Users\HP\Downloads\archive\diabetes.csv")
```

```python
dataset.head(3)
```

```python
x=dataset.iloc[:,:-1]
y=dataset["Outcome"]
```

```python
x.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
fs=SequentialFeatureSelector(lr,k_features=5,forward=True)
fs.fit(x,y)
```

```python
fs.feature_names
```

```python
fs.k_feature_names_
```

```python
fs.k_score_
```

```python
# backward elimination 
```

```python
fs=SequentialFeatureSelector(lr,k_features=5,forward=False)
fs.fit(x,y)
```

```python
fs.feature_names
```

```python
fs.k_feature_names_
```

```python
fs.k_score_
```
