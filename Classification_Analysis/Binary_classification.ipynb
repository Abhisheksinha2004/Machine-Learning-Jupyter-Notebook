```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv(r"C:\Users\HP\Downloads\Social_Network_Ads.csv")
dataset.head(3)
```

```python
dataset.drop(columns=["EstimatedSalary"],inplace=True)
dataset.head(3)
```

```python
sns.scatterplot(x="Age",y="Purchased",data=dataset)
plt.show()
```

```python
x=dataset[["Age"]]
y=dataset[["Purchased"]]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.predict([[40]])
```

```python
sns.scatterplot(x="Age",y="Purchased",data=dataset)
sns.lineplot(x="Age",y=lr.predict(x),data=dataset,color="red")
plt.show()
```
