```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
dataset=pd.read_csv(r"C:\Users\HP\Downloads\archive (6)\Study Hour(Linear Regression).csv")
dataset.head(3)
```

```python
plt.figure(figsize=(5,3))
sns.scatterplot(x="Hours",y="Scores",data=dataset)
plt.show()
```

```python
dataset.isnull().sum()
```

```python
x=dataset[["Hours"]]
y=dataset[["Scores"]]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)  #train model in y=mx+c
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
lr.score(x_test,y_test)*100
```

```python
#2.5	 21
```

```python
lr.predict([[2.5]])
```

```python
#y=mx+c,m=9.68207815,c=2.82689235
```

```python
9.68207815*2.5+2.82689235
```

```python
y_prd=lr.predict(x)
```

```python
plt.figure(figsize=(5,4))
sns.scatterplot(x="Hours",y="Scores",data=dataset,label="org")
plt.plot(dataset["Hours"],y_prd,c="red",label="predict line")
plt.legend(["org","predict"])
plt.savefig("predict.jpg")
plt.show()
```
